### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

```python

import pandas as pd
chipo = pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv', sep = '\t')
```

### Step 3. Assign it to a variable called chipo.

### Step 4. How many products cost more than $10.00?

```python

# Convert item price to float
chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)

# Filter for products costing more than $10.00
expensive_products = chipo[chipo['item_price'] > 10.00]

# Count the number of expensive products
num_expensive_products = len(expensive_products)

print(f"{num_expensive_products} products cost more than $10.00")
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python

chipo['item_price'] = chipo['item_price'].str.replace('$', '', regex=False).astype(float)
price_per_item = chipo[['item_name', 'item_price']]
price_per_item
```

### Step 6. Sort by the name of the item

```python
chipo_sorted = chipo.sort_values(by='item_name')
chipo_sorted
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_item = chipo.loc[chipo['item_price'].idxmax()]

# Get the quantity of the most expensive item
quantity_most_expensive = most_expensive_item['quantity']

print(f"The quantity of the most expensive item ordered was: {quantity_most_expensive}")
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
veggie_salad_bowl_orders = chipo[chipo['item_name'] == 'Veggie Salad Bowl']
num_veggie_salad_bowls = len(veggie_salad_bowl_orders)
print(f"A Veggie Salad Bowl was ordered {num_veggie_salad_bowls} times.")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda_orders = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]
num_multiple_canned_sodas = len(canned_soda_orders)
print(f"Someone ordered more than one Canned Soda {num_multiple_canned_sodas} times.")
```
